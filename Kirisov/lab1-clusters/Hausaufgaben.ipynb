```python
import numpy as np
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans

%matplotlib inline
```

```python
papers = fetch_20newsgroups(subset='all', shuffle=True)
```

```python
print '\n'.join(papers.target_names)
```

```python
print 'files:', len(papers.data)
```

```python
data = TfidfVectorizer().fit_transform(papers.data)
```

```python
kmeans = KMeans(n_clusters=len(papers.target_names), n_init=1).fit(data)
```

```python
leet2d = TruncatedSVD(2).fit_transform(data)
leet3d = TruncatedSVD(3).fit_transform(data)
```

```python
plt.figure(figsize = (15, 15))
plt.scatter(leet2d[:,0], leet2d[:,1], c=kmeans.labels_, alpha=0.9)
plt.axis([0, 0.8 , -0.25,0.3])
```

```python
x = plt.figure(figsize = (13, 13)).gca(projection='3d')
x.scatter(leet3d[:,0], leet3d[:,1], leet3d[:,2], c=kmeans.labels_)
plt.axis([0, 0.8 , -0.25,0.3])
```
