## Кластеризация текстов в корпусе 20newsgroups

```python
from sklearn.datasets import fetch_20newsgroups
from sklearn.feature_extraction.text import TfidfVectorizer
import sklearn.cluster
import sklearn.decomposition
import matplotlib.pyplot
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline
corpus = fetch_20newsgroups()
```

```python
tfidf = TfidfVectorizer()
tfidf_matrix = tfidf.fit_transform(corpus.data) # Transform corpus into tf-idf matrix.
```

```python
PCA = sklearn.decomposition.TruncatedSVD(2)
principal_components = PCA.fit_transform(tfidf_matrix)
```

## Проекция на первые две главные компоненты 

### 2 кластера k-means

```python
labels = sklearn.cluster.k_means(tfidf_matrix, 2)[1]
matplotlib.pyplot.figure(figsize = (15, 15))
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 0], principal_components[:, 1][labels == 0], color = 'red', alpha = 0.5)
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 1], principal_components[:, 1][labels == 1], color = 'blue', alpha = 0.5)
matplotlib.pyplot.show()
```

### 3 кластера k-means

```python
labels = sklearn.cluster.k_means(tfidf_matrix, 3)[1]
matplotlib.pyplot.figure(figsize = (15, 15))
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 0], principal_components[:, 1][labels == 0], color = 'red', alpha = 0.5)
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 1], principal_components[:, 1][labels == 1], color = 'blue', alpha = 0.5)
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 2], principal_components[:, 1][labels == 2], color = 'green', alpha = 0.5)
matplotlib.pyplot.show()
```

### 3 кластера k-means

```python
labels = sklearn.cluster.k_means(tfidf_matrix, 4)[1]
matplotlib.pyplot.figure(figsize = (15, 15))
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 0], principal_components[:, 1][labels == 0], color = 'red', alpha = 0.3)
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 1], principal_components[:, 1][labels == 1], color = 'blue', alpha = 0.3)
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 2], principal_components[:, 1][labels == 2], color = 'green', alpha = 0.3)
matplotlib.pyplot.scatter(principal_components[:, 0][labels == 3], principal_components[:, 1][labels == 3], color = 'yellow', alpha = 0.3)
matplotlib.pyplot.show()
```
