```python
import numpy as np
from sklearn.datasets import fetch_20newsgroups_vectorized
from sklearn.decomposition import TruncatedSVD
from sklearn.cluster import KMeans
from sklearn.cluster import SpectralClustering
import plotly.plotly as py
import plotly.graph_objs as go
```

```python
data = fetch_20newsgroups_vectorized(subset='all')
print data
```

```python
dataTo2d = TruncatedSVD(2).fit_transform(data.data)
dataTo3d = TruncatedSVD(3).fit_transform(data.data)
```

```python
kmeansLabels = KMeans(n_clusters=len(data.target_names), n_init=1).fit_predict(data.data)
```

```python
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D
%matplotlib inline

plt.figure(figsize = (20, 20))
plt.scatter(dataTo2d[:,0], dataTo2d[:,1], c=kmeansLabels, alpha=0.75)
plt.axis([0, 1 , -0.5, 0.5])
```

```python
x = plt.figure(figsize = (20, 20)).gca(projection='3d')
x.scatter(dataTo3d[:,0], dataTo3d[:,1], dataTo3d[:,2], c=kmeansLabels, alpha=0.75)
plt.axis([0, 1 , -0.5,0.5])
```
