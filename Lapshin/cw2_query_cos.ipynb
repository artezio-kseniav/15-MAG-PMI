```python
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.datasets import fetch_20newsgroups
from sklearn.metrics.pairwise import cosine_similarity
dataset = fetch_20newsgroups(subset='all')
```

```python
def make_search(value):
    tfidfVectorizer = TfidfVectorizer(stop_words='english', analyzer='word')
    tfidf_index = tfidfVectorizer.fit_transform(dataset.data)

    query = tfidfVectorizer.transform([value])
    words = value.split()
    
    similarity = {}
    index = 0
    for document in tfidf_index:
        similarity[dataset.data[index]] = cosine_similarity(document, query)[0][0]
        index += 1

    res = sorted(similarity.values(), reverse=True)

    for index in range(10):
        text = similarity.keys()[similarity.values().index(res[index])]
        lines = text.split('\n')
        print 'Cosine value -',res[index],'\n'

        for index in range(10):
            print lines[index]
        
        print '+++++++++++++++++++++++++++++++++++++++++\n'
        print '+++++++++++++++++++++++++++++++++++++++++\n'
        print '+++++++++++++++++++++++++++++++++++++++++\n'
```

```python
make_search("Christ")
```
